```python
log_ni = np.array([7, 7.3, 7.7, 8, 8.3, 8.7, 9, 9.3, 9.7, 10])
log_f_ni = np.array([4.12, 4.45, 4.33, 4.14, 3.91, 3.62, 3.37, 3.04, 2.57, 2.21])
```

```python
plt.figure(figsize=[8,6])
plt.plot(log_ni, log_f_ni, 'bo') 
plt.xlabel(r'log $\nu \,(\mathrm{Hz})$')
plt.ylabel(r'log $F_{\nu} \,(\mathrm{Jy})$')
plt.grid()
plt.tight_layout()
#plt.savefig('7')
```

```python
def curva(alpha, log_ni, b):
    return -alpha* log_ni + np.log10(b)

param, param_cov = curve_fit(curva, log_ni, log_f_ni)
  
print(f'coeficientes: {param}')
print(f'erro: {param_cov}')

alpha = param[0]
b = param[1]

fit = -alpha* log_ni + np.log10(b)
```

```python
plt.figure(figsize=[8,6])
plt.plot(log_ni, log_f_ni, 'bo') 
plt.plot(log_ni, fit, 'k-', label=r'$F_{\nu} \propto \nu^{0.71}$')
plt.xlabel(r'log $\nu \,(\mathrm{Hz})$')
plt.ylabel(r'log $F_{\nu} \,(\mathrm{Jy})$')
plt.grid()
plt.legend()
plt.tight_layout()
#plt.savefig('7')
```
